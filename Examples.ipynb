### 1. Create a circle stl file 
#### Excercise: Create and display a circle:

* Step: 1.0: Import the required library:
    - PySTL
* Step: 2.0: Create circle from the Shape class
* Step:3.0: Set the desired attributes on the circle, here we set:
    - radius = 10
    - resolution = 20 (the number of points to be used to create the circle. The more the finer
    the circle.)
* Step 4.0: Call the create method on the shape object (circle here).
* Step 5.0: Visualize the created circle (in this example using pyvista).

**Note: Look for implementation of this below.**

```python
# step 1.0: import PySTL
from PySTL import shapes
# step 2.0: create a circle object from the shaped module
circle=shapes.Cirle()
# set correct mode for display of the shape created -> None uses matplotlib backend
#                                                   -> "pv" uses the pyvista backend to display stl.
circle.mode = "pv"
# set attributes like radius of circle, resolution (how many points to be used to create a circle)
circle.radius = 10.0
circle.resolution = 20
# step 3.0: Once all desired attributes are set, call the create method on the object
circle.create()
# step 4.0: After creating export method on the object can be called to create the stl file
circle.export('Results/circle.stl', 'circle')
# step 5.0: Optionally, the shape can be visulaized by calling the visualize method appropriately
# here is an exaple that uses pyvista backend because the mode is set as pv.
circle.visualize().plot(color='teal', text=f'{circle.name}',jupyter_backend="static")
```
